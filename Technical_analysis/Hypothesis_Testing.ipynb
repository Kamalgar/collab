```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

%matplotlib inline
```

```python

import pandas as pd
import yfinance as yf
import datetime

# Downloading FTSE100 stocks data
aqr2_stocks2 = yf.download(" AZN.L GSK.L SBIN.NS JIOFIN.NS AXS AZN.L  PANW", start=datetime.datetime(1984,1, 1),
                             end=datetime.datetime(2021,1,1), group_by='tickers')

# Save DataFrame to a pickle file
aqr2_stocks2.to_pickle('aqr2_stocks2.pkl')

```

```python
aqr2_idx_to_2020 = pd.read_pickle("aqr2_stocks2.pkl")

aqr2_idx_to_2020
```

```python
# aqr_idx_to_2020['Close'].plot(grid = True)

```

```python
def aqr2_idx_to_2020_plot():
    aqr2_idx_to_2020['Close'].plot(grid = True)
    sns.set(rc={'figure.figsize':(20, 10)})
    plt.axvspan('1984','1989',color='r',alpha=.5)
    plt.axvspan('2008','2010',color='r',alpha=.5)
    plt.axvspan('2020','2021',color='r',alpha=.5)
    labs = mpatches.Patch(color='red',alpha=.5, label="Black Monday, 2008 Crash & Covid-19 Crash")
    plt.legend(handles=[labs], prop={"size":15},  bbox_to_anchor=(0.4, 0.1), loc='upper center', borderaxespad=0.)
    plt.title('Close Price for FTSE 100 stocks', color = 'black', fontsize = 20)
    plt.xlabel('Year', color = 'black', fontsize = 15)
    plt.ylabel('Close Price', color = 'black', fontsize = 15)
    plt.show();

aqr2_idx_to_2020_plot()
```
